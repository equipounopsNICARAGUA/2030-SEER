```python
merged_data = df_excel.merge(coord_excel, on='country')
# Use the drop_duplicates method to eliminate duplicates based on 'Year' within each country
merged_data = merged_data.drop_duplicates(subset=['country', 'Year'])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Assuming you've already filtered the DataFrame and selected columns as mentioned in your code

# Split the data into training (2000-2010) and testing (2011-2021) sets
train_data = merged_data[(merged_data['Year'] >= 2000) & (merged_data['Year'] <= 2010)]
test_data = merged_data[(merged_data['Year'] >= 2011) & (merged_data['Year'] <= 2021)]

# Handle missing values in the 'Pop_elec' column
train_data.dropna(subset=['Pop_elec'], inplace=True)
test_data.dropna(subset=['Pop_elec'], inplace=True)

# Standardize the numerical feature using the same scaler for both train and test data
scaler = StandardScaler()
train_data[['Year']] = scaler.fit_transform(train_data[['Year']])
test_data[['Year']] = scaler.transform(test_data[['Year']])  # Use the same scaler as for train_data

# Create a list of unique 'country'
unique_countries = train_data['country'].unique()

# Initialize an empty DataFrame to store predictions
all_predictions = pd.DataFrame()

# Train and predict for each 'country'
for country in unique_countries:
    # Filter data for the current 'country'
    train_X = train_data[train_data['country'] == country][['Year']]
    train_y = train_data[train_data['country'] == country]['Pop_elec']
    
    # Train a linear regression model for the current region
    model = LinearRegression()
    model.fit(train_X, train_y)

    # Create prediction data for 2022 to 2030 for the current region
    future_years = list(range(2022, 2031))
    prediction_data = pd.DataFrame({'Year': future_years})

    # Standardize the 'Year' feature for prediction using the same scaler
    prediction_data[['Year']] = scaler.transform(prediction_data[['Year']])

    # Make predictions for the current region
    predictions = model.predict(prediction_data)

    # Create a DataFrame with the predictions for the current region
    region_predictions = pd.DataFrame({'country': country, 'Year': future_years, 'Predicted_Pop_elec': predictions})

    # Append the region's predictions to the overall predictions DataFrame
    all_predictions = pd.concat([all_predictions, region_predictions], ignore_index=True)

# Print or use the 'all_predictions' DataFrame for the predicted 'Pop_elec' for 2022 to 2030 for each 'country'
print(all_predictions)



```

```python
# Export to JSON
all_predictions.to_json('access_to_electricity_pred.json', orient='records')

```
